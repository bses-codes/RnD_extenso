```python
import mysql.connector as connection
import pandas as pd
from kafka import KafkaProducer
from datetime import datetime
import time
import random
import numpy as np
```

```python
conn = connection.connect(
    host="Bisheshs-MacBook-Pro.local", 
    database = 'client_rw',
    user='root', 
    password='12345678')
query = "Select * from fc_transaction_base"
df = pd.read_sql(query,conn)
df
```

```python
KAFKA_TOPIC_NAME_CONS = "FC_Transaction_Base"
KAFKA_BOOTSTRAP_SERVERS_CONS = 'localhost:9092'

if __name__ == "__main__":
    print("Kafka Producer Application Started ... ")

    kafka_producer_obj = KafkaProducer(bootstrap_servers=KAFKA_BOOTSTRAP_SERVERS_CONS,
                                       value_serializer=lambda x: x.encode('utf-8'))
    
  
    df['order_id'] = np.arange(len(df))

    
    data = df.to_dict(orient="records")
       

    message_list = []
    message = None
    for message in data:
        
        message_fields_value_list = []
               
        message_fields_value_list.append(message["order_id"])
        message_fields_value_list.append(message["account_number"])
        message_fields_value_list.append(message["branch"])
        message_fields_value_list.append(message["transaction_code"])

        message = ','.join(str(v) for v in message_fields_value_list)
        print("Message Type: ", type(message))
        print("Message: ", message)
        kafka_producer_obj.send(KAFKA_TOPIC_NAME_CONS, message)
        time.sleep(1)


    print("Kafka Producer Application Completed. ")
```
